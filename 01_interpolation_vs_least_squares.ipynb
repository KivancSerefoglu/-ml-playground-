```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m= 6
x=np.random.randn(m)
noise=np.random.randn(m)
y= 1.2*(np.sin(2*x))+0.5+noise

plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of x vs y')
plt.grid()
plt.show()
```

# Goal: Find a polynomial of degree m − 1 = 5 that goes through these points.

```python
X_tilde = np.vstack([x**p for p in range(m)]).T

X_tilde_inv = np.linalg.inv(X_tilde)
w = X_tilde_inv @ y
x_grid=np.linspace(min(x),max(x))
X_tilde_grid=np.vstack([x_grid**p for p in range(m)]).T
y_grid=X_tilde_grid @ w
plt.plot(x_grid,y_grid,color='red',label='Fitted curve')
plt.scatter(x,y,color='blue',label='Data points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Fit to Data')
plt.legend()
plt.grid()
plt.show()
```

# GOAL: fit a degree 1 polynomial to 6 points

```python
X_tilde_one_degree = np.vstack([x**p for p in range(2)]).T
X_tilde_one_degree_inv = np.linalg.pinv(X_tilde_one_degree)
w_one_degree = X_tilde_one_degree_inv @ y

X_one_degree_grid = np.linspace(min(x),max(x))
X_tilde_one_degree_grid = np.vstack([X_one_degree_grid**p for p in range(2)]).T
y_one_degree_grid = X_tilde_one_degree_grid @ w_one_degree

plt.plot(X_one_degree_grid,y_one_degree_grid,color='green',label='Linear Fit')
plt.plot(x_grid,y_grid,color='red',label='Polynomial Fit')
plt.scatter(x,y,color='blue',label='Data points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear vs Polynomial Fit')
plt.legend()
plt.grid()
plt.show()
```
