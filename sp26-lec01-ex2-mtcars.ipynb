```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('mtcars.csv', index_col=0)

x = df['wt'].values
idx_min = np.argmin(x)
y = df['mpg'].values

y[idx_min] += 100

X = np.vstack([x**p for p in range(2)]).T

theta = np.linalg.inv(X.T @ X) @ X.T @ y
b, w = theta

x_grid = np.linspace(np.min(x),np.max(x))

plt.figure(figsize = (5,3))
plt.scatter(x, y)
plt.plot(x_grid, b + w*x_grid, "r--")
plt.xlabel('weight')
plt.ylabel('miles per gallon')
plt.ylim(5,40)
plt.tight_layout()
```

```python

def plot_with_broken_axis(x, y, b, w, slack=1.0, figsize=(5, 3.8)):
    """
    Plot data and a fitted line with a broken y-axis to show an outlier.

    Parameters
    ----------
    x : array-like
        Input feature values.
    y : array-like
        Response values (already including outlier).
    b, w : float
        Parameters for the line: y = b + w*x.
    slack : float, optional
        Vertical padding around the outlier window.
    figsize : tuple, optional
        Figure size.

    Returns
    -------
    fig : matplotlib.figure.Figure
        The figure object.
    """
    
    idx_min = np.argmin(x)
    y_out = y[idx_min]  # assume first entry is the outlier
    x_grid = np.linspace(np.min(x), np.max(x), 200)

    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=figsize,gridspec_kw={'height_ratios': [1, 5]})
    fig.subplots_adjust(hspace=0.05)

    # Plot on both axes
    for ax in (ax1, ax2):
        ax.scatter(x, y)
        ax.plot(x_grid, b + w * x_grid, "r--")

    # Top window = outlier
    ax1.set_ylim(y_out - slack, y_out + slack)
    # Bottom window = normal range
    ax2.set_ylim(5, 40)

    ax2.set_xlabel('weight')
    ax2.set_ylabel('miles per gallon')

    # Hide the spines between axes
    ax1.spines.bottom.set_visible(False)
    ax2.spines.top.set_visible(False)
    ax1.xaxis.tick_top()
    ax1.tick_params(labeltop=False)
    ax2.xaxis.tick_bottom()

    # Cut marks
    d = .5
    kwargs = dict(marker=[(-1, -d), (1, d)], markersize=12,
                  linestyle="none", color='k', mec='k', mew=1, clip_on=False)
    ax1.plot([0, 1], [0, 0], transform=ax1.transAxes, **kwargs)
    ax2.plot([0, 1], [1, 1], transform=ax2.transAxes, **kwargs)

    return fig
```

```python
fig = plot_with_broken_axis(x, y, b, w)

```

# Gradient descent with square loss

```python
theta = np.array([1.0,-1.0])

lr = 0.01
for i in range(10000):
    gradient = np.zeros_like(X[0])
    for j in range(X.shape[0]):
        gradient += -2 * ( y[j] - np.dot(X[j], theta)) * X[j] 
    gradient /= X.shape[0]
    theta -= (lr)*gradient
    
b, w = theta
fig = plot_with_broken_axis(x, y, b, w)

```

#  Gradient descent with absolute value loss  

```python
theta = np.array([1.0,-1.0])

lr = 0.02
for i in range(10000):
    gradient = np.zeros_like(X[0])
    for j in range(X.shape[0]):
        # GD with abs value loss
        gradient += - np.sign(y[j]-np.dot(X[j], theta)) * X[j]
    gradient /= X.shape[0]
    theta -= (lr)*gradient
    
b, w = theta
fig = plot_with_broken_axis(x, y, b, w)
```

# Gradient descent with the square loss (2 polynomial)

```python
X = np.vstack([x**p for p in range(3)]).T

theta = np.linalg.inv(X.T @ X) @ X.T @ y
b, w1, w2 = theta

x_grid = np.linspace(np.min(x),np.max(x))

plt.figure(figsize = (5,3))
plt.scatter(x, y)
plt.plot(x_grid, b + w1*x_grid + w2*x_grid**2, "r--")
plt.xlabel('weight')
plt.ylabel('miles per gallon')
plt.ylim(5,40)
plt.tight_layout()
```

```python

X = np.vstack([x**p for p in range(3)]).T


theta = np.array([1.0,-1.0, 0.0])

lr= 0.005
for i in range(10000):
    gradient = np.zeros_like(theta)
    for j in range(X.shape[0]):
        gradient += -2 * (y[j]-np.dot(X[j], theta)) * X[j]
    gradient /= X.shape[0]
    theta -= (lr)*gradient

b, w1, w2 = theta   
plt.figure(figsize = (5,3))
plt.scatter(x, y)
plt.plot(x_grid, b + w1*x_grid + w2*x_grid**2, "r--")
plt.xlabel('weight')
plt.ylabel('miles per gallon')
plt.ylim(5,40)
plt.tight_layout()
print("mse: ", np.mean((y - (b + w1*x + w2*x**2))**2))
print(theta)

```

# Gradient descent with the absolute loss (2 polynomial)

```python
X = np.vstack([x**p for p in range(3)]).T

theta = np.array([1.0,-1.0, -1.0])

lr= 0.1
for i in range(10000):
    gradient = np.zeros_like(X[0])
    for j in range(X.shape[0]):
        # GD with abs value loss
        gradient += - np.sign(y[j]-np.dot(X[j], theta)) * X[j]
    gradient /= X.shape[0]
    theta -= (lr)*gradient

b, w1, w2 = theta
plt.figure(figsize = (5,3))
plt.scatter(x, y)
plt.plot(x_grid, b + w1*x_grid + w2*x_grid**2, "r--")
plt.xlabel('weight')
plt.ylabel('miles per gallon')
plt.ylim(5,40)
plt.tight_layout()
print("mse loss:", np.mean((y - (b + w1*x + w2*x**2))**2))
print(theta)
```
